```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
import sympy as sm

# plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['font.sans-serif'] = ['YouYuan']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

%matplotlib inline
```

```python
x = np.linspace(-5, 5, 200)
y1 = np.e**x-3*x
y2 = np.sin(x)
y3 = np.cos(x)
y4 = np.log(x+6)
```

## plot 图像绘制

```python
fig = plt.figure()

plt.subplot(121)
plt.plot(x, y1)
plt.title('原图')

plt.subplot(122)
plt.plot(x, y1)

# 设置Axes标题
plt.title('属性更改后图')

# 设置x, y轴标签
"""
    fontsize: 14
    color: 
"""
plt.xlabel('坐标x', fontsize=22, color='r', )
plt.ylabel('坐标y', verticalalignment='top')


# 设置坐标轴范围
"""
    即定义域与值域 [x1, x2, y1, y2]
"""
plt.axis([0, 5, -10, 100])


# 设置坐标轴分度值
"""
    linear: 线性 <Default>
    log: 对数
    symlog: log10对数
    logit: ???
    
"""
plt.xscale('linear')
plt.yscale('linear')

plt.subplots_adjust(top=1, bottom=0, left=0, right=2, hspace=0.05, wspace=0.25)
plt.show()
```

```python
fig, ax_lst = plt.subplots(1, 3)

# picture 1
ax_lst[0].set_title('图1')
ax_lst[0].plot(x, y1)

# picture 2
ax_lst[1].set_title('图2', loc='left', 
                   fontsize=24,
                   color='red',)
ax_lst[1].plot(x, y1)

# picture 3
ax_lst[2].set_title('图3', loc='right', 
                   fontsize=32,
                   color='y',)
ax_lst[2].plot(x, y1)


fig.subplots_adjust(top=1, right=3)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.axis([0, 5, 0, 5])

# 生成表达式
t = sm.symbols('t')
f = sm.Function('f')
expr_latex = sm.latex(f(t).diff(t, 2) + t**3 -2*t + 4)

# 在 (1, 2) 处标注表达式
ax.text(1, 2, '${}$'.format(expr_latex), fontsize=15)


# 在 (2, 3) 处标注框图-Benjamin142857
ax.text(2, 3, 'Benjamin142857', 
        style='italic',
        fontsize=20,
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})


plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.axis([0, 5, 0, 5])

# 在(0, 0)处添加一句 Benjamin142857
ax.text(5, 5, 'Benjamin142857', verticalalignment='top', horizontalalignment='right')

# 在(2, 3)处添加一句 Benjamin-stella，颜色为pink, 字体大小为26
ax.text(2, 3, 'Benjamin-stella', 
        color='pink',
        fontsize=26,
        style='oblique')

plt.show()
```

## subplot 子图绘制

```python
plt.figure()

# 1.1
plt.subplot(121)
plt.plot(x, y1)
plt.title('1.1')
plt.grid(True)

# 1.2
plt.subplot(122)
plt.plot(x, y2)
plt.title('1.2')
plt.grid(True)

plt.subplots_adjust(top=1, bottom=0, left=0, right=2, hspace=0.05, wspace=0.25)
plt.show()
```

```python
plt.figure()

# 1.1
plt.subplot(221)
plt.plot(x, y1)
plt.title('1.1')
plt.grid(True)

# 1.2
plt.subplot(222)
plt.plot(x, y2)
plt.title('1.2')
plt.grid(True)

# 2.1 
plt.subplot(223)
plt.plot(x, y3)
plt.title('2.1')
plt.grid(True)

# 2.2
plt.subplot(224)
plt.plot(x, y4)
plt.title('2.2')
plt.grid(True)


"""
    plt.subplots_adjust: 调整子图画布
    bottom: 总画布下边界坐标（建议0）
    top: 总画布上边界坐标（建议为子图最大行数）
    left: 总画布左边界坐标（建议0）
    right:  总画布右边界坐标（建议为子图最大列数）
    hspace: 子图高度间隔
    wspace: 子图宽度间隔
"""
plt.subplots_adjust(bottom=0, top=2, left=0, right=2, hspace=0.25, wspace=0.25)
plt.show()
```

```python
fig = plt.figure()
y_lst = [y1, y2, y3]

for i, y in enumerate(y_lst):
    ax = fig.add_subplot(1, 3, i+1)
    ax.plot(x, y)
    ax.grid(True)
    plt.title(str(i+1))


plt.subplots_adjust(top=1, bottom=0, left=0, right=3, hspace=0.05, wspace=0.25)
plt.show()
```
