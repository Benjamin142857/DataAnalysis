```python
import numpy as np
import scipy as sp
from scipy.integrate import quad, dblquad, odeint, nquad
from scipy.optimize import linprog
import matplotlib.pyplot as plt

%matplotlib
```

## Test 1

```python
z = np.array([-2, -4, -3])
stA = np.array([
    [3, 4, 2],
    [2, 1, 2],
    [1, 3, 2]
])
stb = np.array([
    60,
    40,
    80,
])
x_bound = (
    (0, np.inf),
    (0, np.inf),
    (0, np.inf),
)

res = linprog(z, stA, stb, bounds=x_bound, options={'disp': True})
res
```

## Test2

```python
# argmax { z = 4x + 3y } 
z = np.array([-4, -3])

stA = np.array([
    [2, 1],     # A: 2x+y <= 10 
    [1, 1],      # B: x+y <= 8
    [0, 1],      # C: y <= 7
])
stb = np.array([
    10,
    8,
    7,
])
bound = (
    (0, np.inf),
    (0, np.inf),
)

res = linprog(z, stA, stb, bounds=bound, options={'disp': True})
res
```
